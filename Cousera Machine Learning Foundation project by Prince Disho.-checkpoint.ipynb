Course Project to build a model to predict the electrical energy output of a Combined Cycle Power Plant, which uses a combination of gas turbines, steam turbines, and heat recovery steam generators to generate power. We have a set of 9568 hourly average ambient environmental readings from sensors at the power plant which we will use in our model.

The columns in the data consist of hourly average ambient variables:
- Temperature (T) in the range 1.81°C to 37.11°C,
- Ambient Pressure (AP) in the range 992.89-1033.30 milibar,
- Relative Humidity (RH) in the range 25.56% to 100.16%
- Exhaust Vacuum (V) in the range 25.36-81.56 cm Hg
- Net hourly electrical energy output (PE) 420.26-495.76 MW (Target we are trying to predict)

Import the pandas for reading dataframes and numpy for reading n-dimentional array.
Then import dataset to the Jupyter notebook, to start working on it. 

```python
import pandas as pd
import numpy as np

df=pd.read_csv('CCPP_data.csv')
df
```

```python
#Assigning values of X and y 
#y=df.iloc[:,3].values
X=df[['AT']]
y=df[['PE']]
```

Start by linear regresion model where we spliting the dataset into training and test set.

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=3/5,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
print('b0=%.2f'%regressor.intercept_)
print('b1=%.2f'%regressor.coef_)
#print('Intercept',regressor.intercept_)
#print('Coef',regressor.coef_)
```

```python
#We PE as our target values and Absolute pressure as independent feature since it shows strong positive correlation.
```

```python
y_prediction=regressor.predict(X_test)
print(y_prediction)
```

```python
y_test,y_prediction
```

```python
from sklearn.metrics import mean_squared_error
print('Mean_squared_error: %.2f'%mean_squared_error(y_test,y_prediction))
print('Root mean square error: %.2f'%np.sqrt(mean_squared_error(y_test,y_prediction)))
```

```python
#visualizing train data
plt.scatter(X_train,y_train,color='Red')
plt.plot(X_train,regressor.predict(X_train),color=None)
plt.title('Train data')
plt.xlabel('AP')
plt.ylabel('PE')
plt.show()
```
