# Task 3

```python

top_10=updated_data.nlargest(10, ['Confirmed'])
top__10=top_10.drop(top_10.columns[[0,1,2,4]],axis=1)
top__10
```

```python
top__10 = top__10.groupby(df['Country/Region']).sum().sort_values("Confirmed")
top__10
#top_10 = top_10.groupby('Country/Region').sum().apply(lambda x:x.sort_values('Confirmed'))
#top_10
```
