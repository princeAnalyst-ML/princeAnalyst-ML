# DATA PROCESSING

1. BINARIZATION 
2. SCALING 
3.NORMALIZATION 
4.MEAN REMOVAL

```python
from sklearn import preprocessing
```

```python
#BINARIZATION
bds_binarized=preprocessing.Binarizer(threshold=0.5).transform(bds_data)
bds_binarized
```

```python
# MEAN REMOVAL
print('Mean=',bds_data.mean(axis=0))
      
print('Std_devetion',bds_data.std(axis=0))
```

```python
bds_scaled=preprocessing.scale(bds_data)
print('Mean=',bds_scaled.mean(axis=0))
print('Std=',bds_scaled.std(axis=0))
```

```python
# SCALING. Transform feature vectors between (0,1)

data_scaler_minmax=preprocessing.MinMaxScaler(feature_range=(0,1))
data_scaled_minmax=data_scaler_minmax.fit_transform(bds_data)
data_scaled_minmax
```

```python
#NORMALIZATION 
# L1_Normalization.(Least Absolute Deviations). Sum of absolute values targeted to be upto 1.

data_normalized_l1=preprocessing.normalize(bds_data,norm='l1')
print('L1 Normalized'
      ,data_normalized_l1)
```

```python
#L2 Normalization(Least Squares). Modifies values such that sum of squares is always up to 1 in each row.
data_normalized_l2=preprocessing.normalize(bds_data, norm='l2')
print('data_normalize_l2',data_normalized_l2)
```
